```python
import datetime

import plotly.express as px
import pandas as pd
```

```python
data = pd.read_csv('COVID-19_aantallen_gemeente_per_dag.csv', sep=';')
data
```

```python
daily_cases = data.groupby('Date_of_publication')['Total_reported'].sum()
daily_cases
```

```python
rolling_mean = daily_cases.rolling(7).mean()
rolling_mean
```

```python
today = datetime.date.fromisoformat(daily_cases.last_valid_index())
yesteryear = datetime.date(today.year - 1, today.month, today.day)

fig = px.line(daily_cases, range_x=[yesteryear,today], range_y=[0,150000])
fig.show()
```

```python
fig = px.line(rolling_mean, range_x=[yesteryear,today], range_y=[0,150000])
fig.show()
```
